```python
import cv2
import os
import math
import sys
from PIL import Image, ImageDraw, ImageFont
```

```python
#crea le immagini della percentuale di forca usata (quindi 101 immagini dallo 0% al 100%)


# Dimensioni del rettangolo
base = 200  # pixel
altezza = 500  # pixel

# Genera 100 immagini
for percentuale in range(0, 101):
    # Crea un'immagine vuota con uno sfondo bianco
    immagine = Image.new('RGB', (base, altezza), (255, 255, 255))
    draw = ImageDraw.Draw(immagine)

    # Calcola l'altezza del riempimento in base alla percentuale
    altezza_riempimento = int(altezza * percentuale / 100)

    # Calcola le coordinate del rettangolo
    x1 = 0
    y1 = altezza - altezza_riempimento
    x2 = base
    y2 = altezza

    # Imposta il colore di riempimento come verde chiaro
    colore_riempimento = (144, 238, 144)  # Codice RGBA per il verde chiaro

    # Disegna il riempimento verde chiaro con angoli stondati
    bordo = 6
    draw.rounded_rectangle([(x1, y1), (x2, y2)], fill=colore_riempimento, radius=bordo)

    # Disegna il bordo nero con angoli stondati
    draw.rounded_rectangle([(0, 0), (base, altezza)], outline=(0, 0, 0), width=3, radius=bordo)


    # Scrivi la percentuale
    percentuale_testo = f"{percentuale}%"
    #font = ImageFont.truetype("arial.ttf", 30)
    testo_larghezza, testo_altezza = draw.textsize(percentuale_testo)
    testo_x = int(base / 2 - testo_larghezza / 2)
    testo_y = altezza - 80  # Sposta il testo di 30 pixel sopra il fondo

    draw.text((testo_x, testo_y), percentuale_testo, fill=(0, 0, 0))

    # Salva l'immagine
    immagine.save(f'immagini/immagine_{percentuale}.png')
```

```python
fileName="nomeFileLog.TXT"
forkTravel = 150;
detectionsNumber = math.floor(600*4 / (math.pi *9.4) *forkTravel);


def translate(value, leftMin, leftMax, rightMin, rightMax):
    # Figure out how 'wide' each range is
    leftSpan = leftMax - leftMin
    rightSpan = rightMax - rightMin

    # Convert the left range into a 0-1 range (float)
    valueScaled = float(value - leftMin) / float(leftSpan)

    # Convert the 0-1 range into a value in the right range.
    return rightMin + (valueScaled * rightSpan)


image_dir="immagini/"
with open(fileName) as f: values = f.read().split(",")

values = [int(x) for x in values[0:-2]] #also cut out last element ( empty string )
mn=min(values)
values=[translate(int(x),int(mn),int(mn)+detectionsNumber,0,forkTravel) for x in values]


immagini=[f"{image_dir}/immagine_{math.floor(val*100/forkTravel)}.png" for val in values]



# Impostazioni del video
durata_video_input = 60*3
video_name = "out.mp4"
fps = math.floor(len(immagini)/ durata_video_input)  # Frame per secondo
print(fps)
frame = cv2.imread(immagini[0])
height, width, layers = frame.shape


# Crea l'oggetto VideoWriter
video = cv2.VideoWriter(video_name, cv2.VideoWriter_fourcc(*"mp4v"), fps, (width,height))


i=0
# Crea il video frame per frame
for immagine_nome in immagini:
    print(str(i) + " / " + str(len(immagini)))
    frame = cv2.imread(immagine_nome)
    video.write(frame)
    i+=1


# Chiudi il video
video.release()
```
